```python
ra = nw_parse_from_xls('data/rate-testing_2.56mg.xlsx')
ra.set_activate(2.56e-3)
```

```python
from pandas import DataFrame

ra.statis["CRate"] = ra.statis['AvgmCurrent'] / ra.statis['Capacity']

idx = range(114514)

 

dc = ra.statis[ra.statis['Status'] == '恒流放电']['Capacity']
dcr = ra.statis[ra.statis['Status'] == '恒流放电']['CRate'].apply(lambda x: f'{x:2.1f}C')

```

```python
from plotnine import *

(
    ggplot(aes(y=dc, x=range(dc.shape[0]), color=dcr))
    + geom_point()
)
```
